```python
# import BCSE teams
filepath = r"C:\Users\1545966\Documents\Private\bcse_teams.csv"
teams = pd.read_csv(filepath, encoding='mbcs')
```

```python
# import points
filepath = r"C:\Users\1545966\Documents\Private\points.csv"
pts = pd.read_csv(filepath, encoding='mbcs')
```

```python
# clean up name column
df['Name'] = df['Name'].str.strip()
```

```python
# filter out non-bcse
data = df.loc[df['Name'].str.endswith(tuple(teams['team_tag']))].copy()
nb = df.loc[~df['Name'].str.endswith(tuple(teams['team_tag']))].copy()
```

```python
# now extract team name
searchstr = '|'.join(teams['team_tag'])
data['Team'] = data['Name'].str.extract("(" + searchstr + ")", expand=False)
```

```python
# get rider names
x = data['Name'].str.split(expand=True)
```

```python
data['forename'] = x[0]
data['surname'] = x[1]
```

```python
# split into men/women
mn = data.loc[data['Gender'] == 'Men']
wm = data.loc[data['Gender'] == 'Women']
```

```python
# split into categories
a = mn.loc[data['Cat']=='A'].copy().reset_index()
b = mn.loc[data['Cat']=='B'].copy().reset_index()
c = mn.loc[data['Cat']=='C'].copy().reset_index()
d = mn.loc[data['Cat']=='D'].copy().reset_index()

aw = wm.loc[data['Cat']=='A'].copy().reset_index()
bw = wm.loc[data['Cat']=='B'].copy().reset_index()
cw = wm.loc[data['Cat']=='C'].copy().reset_index()
dw = wm.loc[data['Cat']=='D'].copy().reset_index()
```

```python
# allocate points per each category
cats = ['A','B','C','D']
```

```python
# men
dfs = [a,b,c,d]
for i, df in enumerate(dfs):

    df['position'] = df.index+1

    col = cats[i]

    dfs[i] = pd.merge(df, pts[[col]], left_index=True, right_index=True)

    # rename points column
    dfs[i].rename(columns={col: 'Points'}, inplace=True)
    
    # drop unnecessary index column
    dfs[i].drop('index', axis=1, inplace=True)
```

```python
[a,b,c,d] = dfs
```

```python
# women
dfs = [aw,bw,cw,dw]
for i, df in enumerate(dfs):

    df['position'] = df.index+1

    col = cats[i]

    dfs[i] = pd.merge(df, pts[[col]], left_index=True, right_index=True)

    # rename points column
    dfs[i].rename(columns={col: 'Points'}, inplace=True)

    # drop unnecessary index column
    dfs[i].drop('index', axis=1, inplace=True)
```

```python
[aw,bw,cw,dw] = dfs
```
